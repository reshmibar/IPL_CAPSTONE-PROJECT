```python
batsman_stats = merged_df.groupby('Striker').agg(
    total_runs=('runs_scored', 'sum'),
    balls_faced=('runs_scored', 'count'),  # Fix here
    fours=('runs_scored', lambda x: (x == 4).sum()),
    sixes=('runs_scored', lambda x: (x == 6).sum())
).reset_index()

# Calculate derived metrics
batsman_stats['strike_rate'] = (batsman_stats['total_runs'] / batsman_stats['balls_faced']) * 100
batsman_stats['boundary_runs'] = (batsman_stats['fours'] * 4) + (batsman_stats['sixes'] * 6)
batsman_stats['boundary_pct'] = (batsman_stats['boundary_runs'] / batsman_stats['total_runs']) * 100

```

```python
# Filter only legal deliveries (exclude wides and no-balls)
legal_deliveries = merged_df[
    (~merged_df['extras'].isin(['wides', 'noballs'])) | (merged_df['extras'].isnull())
]
```

```python
 #STEP 1: Calculate batsman metrics using legal deliveries only
batsman_stats = legal_deliveries.groupby('Striker').agg(
    total_runs=('runs_scored', 'sum'),
    balls_faced=('runs_scored', 'count'),  # count only legal deliveries
    fours=('runs_scored', lambda x: (x == 4).sum()),
    sixes=('runs_scored', lambda x: (x == 6).sum())
).reset_index()

# STEP 2: Derive strike rate and boundary %
batsman_stats['strike_rate'] = (batsman_stats['total_runs'] / batsman_stats['balls_faced']) * 100
batsman_stats['boundary_runs'] = (batsman_stats['fours'] * 4) + (batsman_stats['sixes'] * 6)
batsman_stats['boundary_percent'] = (batsman_stats['boundary_runs'] / batsman_stats['total_runs']) * 100
```

```python
batsman_stats['strike_rate']
batsman_stats['boundary_percent']
```

```python
# Define thresholds
sr_threshold = batsman_stats['strike_rate'].quantile(0.75)
bp_threshold = batsman_stats['boundary_percent'].quantile(0.75)

# Classification logic
def classify(row):
    if row['strike_rate'] >= sr_threshold and row['boundary_percent'] >= bp_threshold:
        return 'Aggressive'
    elif row['strike_rate'] < sr_threshold and row['boundary_percent'] < bp_threshold:
        return 'Anchor'
    else:
        return 'Balanced'

# Apply classification
batsman_stats['style'] = batsman_stats.apply(classify, axis=1)
```

```python
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

sns.scatterplot(
    data=batsman_stats,
    x='boundary_percent',
    y='strike_rate',
    hue='style',
    palette={'Aggressive': 'red', 'Anchor': 'blue', 'Balanced': 'green'},
    s=100,
    alpha=0.7
)

plt.title('Batting Styles: Boundary % vs Strike Rate', fontsize=16)
plt.xlabel('Boundary Percentage (%)', fontsize=12)
plt.ylabel('Strike Rate', fontsize=12)
plt.legend(title='Batting Style', loc='upper left')
plt.tight_layout()
plt.show()
```

```python
bowling_stats = legal_deliveries.groupby('Bowler').agg(
    balls_bowled=('Ball No', 'count'),
    runs_conceded=('runs_scored', 'sum'),
    dot_balls=('runs_scored', lambda x: (x == 0).sum()),
    wickets=('Player Out', lambda x: x.notnull().sum())  # Correct column name
).reset_index()
```

```python
# Convert balls to overs
bowling_stats['overs'] = bowling_stats['balls_bowled'] / 6

# Economy Rate: Runs conceded per over
bowling_stats['economy'] = bowling_stats['runs_conceded'] / bowling_stats['overs']

# Bowling Average: Runs per wicket
bowling_stats['average'] = bowling_stats['runs_conceded'] / bowling_stats['wickets']
bowling_stats['average'] = bowling_stats['average'].replace([float('inf')], 0)  # Avoid division by 0

# Dot Ball %
bowling_stats['dot_ball_percent'] = (bowling_stats['dot_balls'] / bowling_stats['balls_bowled']) * 100
```

```python
bowling_stats
```

```python
# Filter for consistent bowlers (e.g. min 120 balls bowled)
consistent_bowlers = bowling_stats[bowling_stats['balls_bowled'] >= 120].copy()

# Remove any inf or NaN that might mess up the plot
consistent_bowlers.replace([float('inf')], 0, inplace=True)
consistent_bowlers.dropna(subset=['economy', 'dot_ball_percent', 'average'], inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=consistent_bowlers,
    x='economy',
    y='dot_ball_percent',
    size='wickets',
    hue='average',
    palette='coolwarm',
    sizes=(30, 300),
    alpha=0.7
)
plt.title('Bowling Consistency: Dot Ball % vs Economy Rate')
plt.xlabel('Economy Rate')
plt.ylabel('Dot Ball Percentage')
plt.grid(True)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Add Over Number column
legal_deliveries['Over'] = legal_deliveries['Ball No'].astype(int)

# Categorize overs
def get_phase(over):
    if over <= 6:
        return 'Powerplay'
    elif over <= 15:
        return 'Middle Overs'
    else:
        return 'Death Overs'

legal_deliveries['Phase'] = legal_deliveries['Over'].apply(get_phase)

# Aggregate runs per phase
over_phase_stats = legal_deliveries.groupby('Phase').agg(
    total_runs=('runs_scored', 'sum'),
    total_balls=('Ball No', 'count')
).reset_index()

# Calculate runs per over
over_phase_stats['runs_per_over'] = (over_phase_stats['total_runs'] / over_phase_stats['total_balls']) * 6
```

```python
sns.barplot(data=over_phase_stats, x='Phase', y='runs_per_over', palette='viridis')
plt.title('Average Runs per Over in Different Phases')
plt.ylabel('Runs per Over')
plt.xlabel('Match Phase')
plt.grid(axis='y')
plt.show()
```

```python
# Total runs per match
match_totals = legal_deliveries.groupby(['Match id', 'Venue']).agg(
    total_runs=('runs_scored', 'sum')
).reset_index()
```

```python
# Compute median total score
median_score = match_totals['total_runs'].median()
print(f"Median Match Total: {median_score}")
```

```python
# Categorize matches
match_totals['score_category'] = match_totals['total_runs'].apply(
    lambda x: 'High-Scoring' if x >= median_score else 'Low-Scoring'
)
```
